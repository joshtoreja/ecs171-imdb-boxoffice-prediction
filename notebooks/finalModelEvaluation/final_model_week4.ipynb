```python
# Core libraries
import pandas as pd #data manipulation and tabluar operation
import numpy as np #numerical operations and transformation

# Visualization
import matplotlib.pyplot as plt #base plotting
import seaborn as sns #statistical visualization

# Model selection
from sklearn.model_selection import (
    train_test_split, #train/ test splitting
    StratifiedKFold, #stratified cross validation
    RandomizedSearchCV, #hyperparameter tuning
    cross_val_score #cross validation scoring
)

# Preprocessing
from sklearn.preprocessing import StandardScaler, OneHotEncoder #scaling + categorical encoding
from sklearn.compose import ColumnTransformer #apply different preprocessing to different columns
from sklearn.pipeline import Pipeline #end to end modeling pipeline

# Metrics
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    roc_curve,
    precision_recall_curve,
    confusion_matrix
)

# Model
from xgboost import XGBClassifier #gradient boosted tree model

# Saving
import joblib
import os
```

```python
df = pd.read_csv("../data/processed/tmdb_clean.csv") #load cleaned numeric dataset

print("Dataset shape:", df.shape) # print dataset shape
df.head()#preview of first few rows
```

```python
df.columns
```

```python
# Target
y = df["profitable"] #binary profitability label created earlier (ROI threshold already applied)

# Release year
df["release_year"] = pd.to_datetime(df["release_date"]).dt.year

# Release month
df["release_month"] = pd.to_datetime(df["release_date"]).dt.month

# Seasonal features
df["is_summer"] = df["release_month"].isin([5,6,7]).astype(int) #summer release indicator
df["is_holiday"] = df["release_month"].isin([11,12]).astype(int) #holiday release indicator

# Log transform budget
df["log_budget"] = np.log1p(df["budget"]) #apply log1p to reduce heavy skew in budget distribution

# Runtime bucket
df["long_movie"] = (df["runtime"] > 120).astype(int) #binary indicator for long films

# Numerical features
numerical_features = [
    "log_budget",
    "runtime",
    "release_year",
    "is_summer",
    "is_holiday",
    "long_movie"
]

# Categorical features
categorical_features = [
    "genres",
    "production_companies",
    "production_countries"
]

X = df[numerical_features + categorical_features] #combine into feature matrix

print("X shape:", X.shape)
print("Class balance:\n", y.value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_features),#scale numerical features
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features) #one-hot encode categorical features
    ]
)
```

```python
model = XGBClassifier(
    objective="binary:logistic", #binary classifcation
    eval_metric="logloss", #optimization metric
    random_state=42, #reproducability
    scale_pos_weight=1.2  # mild imbalance adjustment 
)

pipeline = Pipeline([ #combine into pipeline
    ("preprocessor", preprocessor),
    ("model", model)
])
```

```python
#hyperparameter search space
param_grid = {
    "model__n_estimators": [400, 600, 800],# number of bosting trees
    "model__learning_rate": [0.03, 0.05, 0.07], #step size shrinkage
    "model__max_depth": [3, 4, 5], #tree depth
    "model__subsample": [0.7, 0.8], #row sampling
    "model__colsample_bytree": [0.7, 0.8], #feature sampling
    "model__min_child_weight": [1, 3] #regularization
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) # 5-fold stratified cross-validaton

search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_grid,
    n_iter=25,# number of random parameter combinations
    scoring="f1",  # optimize for F1 score
    cv=cv,
    verbose=1,
    n_jobs=-1, # prallel processing
    random_state=42
)

search.fit(X_train, y_train) #perform randomized hyperparameter tuning

print("Best Params:", search.best_params_) #print the best hyperparameters
print("Best CV F1:", search.best_score_)

pipeline = search.best_estimator_ #retrieve the best pipeline
```

```python
cv_auc = cross_val_score( #evaluate ROC-AUC using cross validation
    pipeline,
    X_train,
    y_train,
    cv=cv,
    scoring="roc_auc"
)

print("CV ROC-AUC:", cv_auc.mean())
```

```python
pipeline.fit(X_train, y_train) #train final model on training data
```

```python
y_val_proba = pipeline.predict_proba(X_train)[:, 1] #get predicted probabilities on training data

precision, recall, thresholds = precision_recall_curve(y_train, y_val_proba) # precision recall curve

f1_scores = 2 * (precision * recall) / (precision + recall + 1e-9) #compute F1 at each threshold

#identify threshold maximzing F1
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]

print("Best Threshold:", best_threshold)
```

```python
y_test_proba = pipeline.predict_proba(X_test)[:, 1] #predict probabilities on test set
y_test_pred = (y_test_proba >= best_threshold).astype(int) # applky optimized threshold

#print evaluation metrics
print("TEST Accuracy:", accuracy_score(y_test, y_test_pred))
print("TEST Precision:", precision_score(y_test, y_test_pred))
print("TEST Recall:", recall_score(y_test, y_test_pred))
print("TEST F1:", f1_score(y_test, y_test_pred))
print("TEST ROC-AUC:", roc_auc_score(y_test, y_test_proba))
```
